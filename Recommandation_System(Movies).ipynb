```python
import pandas as pd
import numpy as np
df1=pd.read_csv('tmdb_5000_credits.csv.zip')
df2=pd.read_csv('tmdb_5000_movies.csv.zip')
```

```python
df1.shape,df2.shape
```

```python
movies=df2.merge(df1,on='title')
```

```python
movies.shape
```

```python
movies.info()
```

```python
movies=movies[['genres','movie_id','title','keywords','overview','cast','crew']]
movies
```

```python
movies.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.iloc[0].genres
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
movies.iloc[0].cast
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.iloc[0].cast
```

```python
movies.iloc[0].crew
```

```python
def fetch_director(obj):
    L=[]
    for  i in ast.literal_eval(obj):
        if i['job'] =='Director':
            L.append(i['name'])
            break
        
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.iloc[0].crew
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.iloc[0].overview
```

```python
movies
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'].iloc[0]
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ",'') for i in x])
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
movies['tags']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head(20)
```

```python
new_df.tags.iloc[0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.tags.iloc[0]
```

```python
new_df.tags=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.tags.iloc[0]
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'].iloc[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector).shape
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend("Harry Potter and the Half-Blood Prince")
```
